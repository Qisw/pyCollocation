```python
import numpy as np
import sympy as sym
import seaborn as sn

import models
import orthogonal_collocation
```

```python
# define some variables
t, k, c = sym.symbols('t, k, c')

# define some parameters
alpha = sym.symbols('alpha')
rho, theta = sym.symbols('rho, theta')
g, n, delta = sym.symbols('g, n, delta')

# define functional form for intensive output
f = k**alpha

# define functional form for utility
u = c**(1 - theta) / (1 - theta)

# equation of motion for capital (per unit effective labor)
k_dot = f - c - (g + n + delta) * k

# equation of motion for consumption (per unit effective labor)
mpk = sym.diff(f, k, 1)
epsilon_mu = -(sym.diff(u, c, 2) * c) / sym.diff(u, c)
c_dot = ((mpk - delta - rho - theta * g) / epsilon_mu) * c
rhs = {k: k_dot, c: c_dot}

# define some boundary conditions
k0 = 5.0
kstar = (alpha / (delta + rho + theta * g))**(1 / (1 - alpha))
cstar = f.subs({k: kstar}) - (g + n + delta) * kstar
bcs = {'lower': [k - k0], 'upper': [c - cstar]}

ramsey = models.BoundaryValueProblem(dependent_vars=[k, c],
                                     independent_var=t,
                                     rhs=rhs,
                                     boundary_conditions=bcs)

ramsey_params = {'g': 0.02, 'n': 0.02, 'alpha': 0.15,'theta': 1.05, 'rho': 0.03, 'delta': 0.04}

ramsey_solver = orthogonal_collocation.OrthogonalCollocationSolver(ramsey, ramsey_params)
```

```python
# specify an initial guess
def kstar(g, n, alpha, delta, rho, theta):
    return (alpha / (delta + rho + theta * g))**(1 / (1 - alpha))

def cstar(g, n, alpha, delta, rho, theta):
    return kstar(g, n, alpha, delta, rho, theta)**alpha - (g + n + delta) * kstar(g, n, alpha, delta, rho, theta)

domain = [0, 200]
ts = np.linspace(domain[0], domain[1], 1000)
ks = kstar(**ramsey_params) - (kstar(**ramsey_params) - k0) * np.exp(-ts)
initial_capital_poly = np.polynomial.Chebyshev.fit(ts, ks, 15, domain)

cs = np.log(initial_capital_poly(ks))  # guess that consumption is concave function of capital
initial_consumption_poly = np.polynomial.Chebyshev.fit(ts, cs, 15, domain)
initial_ramsey_coefs = {k: initial_capital_poly.coef, c: initial_consumption_poly.coef}
```

```python
ramsey_solution = ramsey_solver.solve(kind="Chebyshev",
                                      coefs_dict=initial_ramsey_coefs,
                                      domain=domain)
```

```python
ramsey_solution.number_knots = 1000
ramsey_solution.solution.plot(subplots=True, style=['r', 'b'])
```
